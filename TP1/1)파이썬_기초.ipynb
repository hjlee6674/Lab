```python
# if문에 해당되지 않는 경우, else를 이용해 간단하게 출력할 수 있음
# 위와 동일한 코드
for i in range(1,11):      
  if i%3==0:                 
    print('박수')            
  else:                  
    print(i) 
```

```python
# elif함수는 else와 if를 합한 함수
# 위에서의 if 조건문을 만족시키지 않을 경우, 아래의 elif문이 작동됨
score='B'

if score=='A':    # 만약 score가 'A'라면
  print('A')      # 'A'를 출력
elif score=='B':  # score가 'A'가 아니면서 'B'라면
  print('B')      # 'B'를 출력
elif score=='C':  # score가 'A','B',가 아니면서 'C'라면
  print('C')      # 'C'를 출력
else:             # score가 'A','B','C'가 아니라면
  print('F')      # 'F'를 출력
```

```python
# if문과 elif문의 차이

score=65
if score>=80:
  print('A')
elif score>=60:
  print('B')
elif score>=40:
  print('C')
else:
  print('F')
```

```python
score=65             # 65점은 60점 이상이면서 40점 이상이므로 B와 C가 모두 출력됨
if score>=80:
  print('A')
if score>=60:
  print('B')
if score>=40:
  print('C')
else:
  print('F')
```

```python
score=65          # 조건문을 이중으로 사용하면 동일한 결과를 얻을 수 있지만 elif에 비해 연산이 느림
if score>=80:
  print('A')
elif 80>score>=60:
  print('B')
elif 60>score>=40:
  print('C')
else:
  print('F')
```

```python
step=0             # while은 조건이 True인 경우 작동, False인 경우 중지
series_sum=0
while step<10:
  series_sum=series_sum+step
  step=step+1
print(series_sum)          # 0~9까지의 합이 출력, step=10인 경우는 계산되지 않았음
```

```python
n=0
while True:   # 조건문을 잘못 작성하면 코드가 무한히 돌아갈 수 있음
  n=n+1       # 중지 버튼을 눌러서 계산 중지
```

```python
for i in range(10):  # if ... break를 사용해 반복문을 탈출할 수 있음
  print(i)
  if i>5:              # i=6일 때 조건문이 정지가 됨
    break
```

```python
# 팩토리얼 연산하기
n=5     
result=1
while n>0:     # n=5,4,3,2,1까지 루프가 돌아감 (n=0일때 정지)
  result=result*n
  n=n-1
print(result)
```

```python
# 소수 찾기 알고리즘
prime_list=[]
for i in range(2,100):  # 2부터 99까지의 i에 대해
  devisors=[]           # 약수를 모두 구한다
  for j in range(1,i+1):  # 1~i까지의 수를 모두 나누어서
    if i%j==0:          # 나머지가 0이 되면 약수에 포함
      devisors.append(j)
  if len(devisors)==2:  # 약수가 2개이면 소수
    prime_list.append(i)
print(prime_list)
```
